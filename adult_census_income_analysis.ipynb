# Handling ? in the Dataset

```python
df['workclass'].value_counts()
```

```python
columns_to_count = ['age', 'workclass', 'fnlwgt', 'education', 'education.num',
                    'marital.status', 'occupation', 'relationship', 'race', 'sex',
                    'capital.gain', 'capital.loss', 'hours.per.week', 'native.country',
                    'income']

for col in columns_to_count:
    print(df[col].value_counts())
    print()  # Optional: Add a blank line between each column's value counts
```

#### columns : Workclass , Occupation and native- coutry has ?

#### The occurence of ? in </br> workclass is : 1836 </br> Ocuupation is : 1843 </br> native- country is :  583

#### we can not drop the ? in either of the column since the ocuurences is high , we can give ? a new feature name or replace it with highest occurence variable called mode

## Replacing values with mode

```python
df['workclass'] = df['workclass'].replace('?','Private')
```

```python
df['occupation'] = df['occupation'].replace('?','Prof-specialty')
```

```python
df['native.country'] = df['native.country'].replace('?','United-States')
```

# Feature Enginnering 

```python
df['education']=df['education'].replace(['Masters'],'graduate')
df['education']=df['education'].replace(['Prof-school'],'training')
df['education']=df['education'].replace(['Doctorate'],'doctrate')
df['education']=df['education'].replace(['Assoc-voc','Assoc-acdm','Some-college'],'college/associate')
df['education']=df['education'].replace(['11th','12th'],'highschool')
df['education']=df['education'].replace(['Bachelors'],'undergraduate')
df['education']=df['education'].replace(['HS-grad'],'highschoolgraduate')
df['education']=df['education'].replace(['Preschool'],'preschool')
df['education']=df['education'].replace(['1st-4th','5th-6th','7th-8th','9th','10th'],'school')  
```

```python
df['marital.status']=df['marital.status'].replace(['Married-civ-spouse','Married-spouse-absent','Married-AF-spouse'],'married')
df['marital.status']=df['marital.status'].replace( ['Never-married'],'unmarried')
df['marital.status']=df['marital.status'].replace( ['Separated','Divorced'],'sperated/divorced')
df['marital.status']=df['marital.status'].replace (['Widowed'],'widowed')

```
